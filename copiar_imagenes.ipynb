```python
import os
import json
import shutil
import pandas as pd
```

```python
json_path ='/Users/tonivila/Library/CloudStorage/GoogleDrive-toni.vila@thevallians.com/Unidades compartidas/The Valley - TFM/Experimento_1/df_clean_run1.json'

df = pd.read_json(json_path)
df.head()
```

```python
def duplicates_finder(label, dir_ori, dir_dest):
    pic_ids = list(df[df['label'] == label].photo_id)

    for i in pic_ids:
        try:
            shutil.copy(dir_ori + '/' + i + '.jpg', dir_dest + label)

        except:
            pass
```

```python
origin = '/Users/tonivila/Downloads/imagenes_negocios'
destination = '/Users/tonivila/Library/CloudStorage/GoogleDrive-toni.vila@thevallians.com/Unidades compartidas/The Valley - TFM/imagenes_seleccionadas'

labels = ['food', 'inside', 'drink', 'ouside', 'menu']

for i in labels:
    duplicates_finder(i, 
                    origin,
                    destination)
```
